## 0-Shot Prompting

## K-Shot Prompting

````python
def add(a, b):
    """Adds two numbers.

    Args:
        a: first number
        b: second number

    Returns:
        The sum of both numbers.
    """
    return a + b
````

````python

examples = [inspect.getsource(add)]
````

````python
user_message_template_string = \
"""Write a docstring for the following function:
Examples:

{% for example in examples %}
{{ example }}
{% endfor %}

Function:
{{ function }}
"""

user_message_template = jinja2.Template(user_message_template_string, undefined=jinja2.StrictUndefined)
print(backend(messages=[{"role": "user", "content": user_message_template.render(function=function, examples=examples)}]).choices[0].message.content)
````

## Chain-of-Thought Prompting

````python
user_message_template_string = \
"""Write a docstring for the following function. Please think step-by-step before giving an answer.

{{ function }}
"""

user_message_template = jinja2.Template(user_message_template_string, undefined=jinja2.StrictUndefined)
print(backend(messages=[{"role": "user", "content": user_message_template.render(function=function)}]).choices[0].message.content)
````
