## Chat Implementation

````python
from IPython import display as ipydisplay

from hslu.dlm03.common import chat as chat_lib
from hslu.dlm03.common import types
from hslu.dlm03.common.displays import ipython_display
````

````python
chat_display = ipython_display.IPythonChatDisplay()
chat_display.show()
````

````python
chat = chat_lib.Chat(observers=[chat_display])
while True:
    user_input = input()
    ipydisplay.clear_output()
    user_message = types.UserMessage(content=user_input, role="user")
    chat.append(user_message)
    response = backend.generate(chat)
    assistant_message = types.AssistantMessage(content=response.choices[0].message.content, role="assistant")
    chat.append(assistant_message)
````
